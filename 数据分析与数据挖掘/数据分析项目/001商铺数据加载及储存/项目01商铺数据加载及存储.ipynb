【项目01】  商铺数据加载及存储

要求：
1、成功读取“商铺数据.csv”文件
2、解析数据，存成列表字典格式：[{'var1':value1,'var2':value2,'var3':values,...},...,{}]
3、数据清洗：
① comment，price两个字段清洗成数字
② 清除字段缺失的数据
③ commentlist拆分成三个字段，并且清洗成数字
4、结果存为.pkl文件

```python
# 加载使用的库
import pandas as pd
import re
import pickle
```

```python
# 读取数据
df = pd.read_csv('shangpushuju.csv')
```

```python
df.head()
```

```python
#函数功能： 将缺失值标为"数据缺失"
def label_Nan(lst):
    i = 0
    for s in lst:
        if s == '':
            lst[i] = "数据缺失"
        i+=1
```

```python
#使用正则表达式，从string中提取非数字内容，并将其替代为""
comment = [re.sub("\D","",s) for s in df['comment']]
label_Nan(comment)
price = [re.sub("\D","",s) for s in df['price']]
label_Nan(price)
```

```python
commentlist = []
for s in df['commentlist']:
    if isinstance(s,str):
        commentlist.append(re.findall(r"\d+\.?\d*",s))
    else:
        commentlist.append("数据缺失")
```

```python
# 函数功能：将comment、price、commentlist里面的数值转换成int
def str2int(lst):
    i = 0
    for s in lst:
        if s == '数据缺失':
            pass
        elif type(s) == str:
            lst[i] = int(s)
        elif type(s) == list:
            lst[i] = list(map(float,s))
        i += 1
```

```python
str2int(comment)
str2int(price)
str2int(commentlist)
```

```python
#将数据存储为元素是字典的列表
data_of_store = []
for i in range(len(price)):
    dic = {}
    dic['classify'] = df['classify'][i]
    dic['name'] = df['name'][i]
    dic['comment'] = comment[i]
    dic['star'] = df['star'][i]
    dic['price'] = price[i]
    dic['address'] = df['address'][i]
    dic['commentlist'] = commentlist[i]
    data_of_store.append(dic)
```

```python
data_of_store
```

```python
# 将数据存储在pickle中
output = open('data_of_store.pkl','wb')
pickle.dump(data_of_store, output)
output.close()
```
