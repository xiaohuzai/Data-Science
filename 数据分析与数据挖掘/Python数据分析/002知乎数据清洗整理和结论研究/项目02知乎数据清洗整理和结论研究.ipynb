```python
'''
【项目02】  知乎数据清洗整理和结论研究

作业要求：
1、数据清洗 - 去除空值
要求：创建函数
提示：fillna方法填充缺失数据，注意inplace参数

2、问题1 知友全国地域分布情况，分析出TOP20
要求：
① 按照地域统计 知友数量、知友密度（知友数量/城市常住人口），不要求创建函数
② 知友数量，知友密度，标准化处理，取值0-100，要求创建函数
③ 通过多系列柱状图，做图表可视化
提示：
① 标准化计算方法 = (X - Xmin) / (Xmax - Xmin)
② 可自行设置图表风格

3、问题2 知友全国地域分布情况，分析出TOP20
要求：
① 按照学校（教育经历字段） 统计粉丝数（‘关注者’）、关注人数（‘关注’），并筛选出粉丝数TOP20的学校，不要求创建函数
② 通过散点图 → 横坐标为关注人数，纵坐标为粉丝数，做图表可视化
③ 散点图中，标记出平均关注人数（x参考线），平均粉丝数（y参考线）
提示：
① 可自行设置图表风格

'''
```

```python
from pylab import mpl
mpl.rcParams['font.sans-serif'] = ['SimHei']
mpl.rcParams['axes.unicode_minus'] = False           # 解决保存图像是负号'-'显示为方块的问题
```

```python
import pandas as pd
```

```python
'''
1、数据清洗，去除空值
'''
def clean_data(file):
    data = pd.read_csv(file,engine = 'python')
    data.fillna("缺失数据",inplace = True)
    return data
#清洗数据
data = clean_data("知乎_201701.csv")
```

```python
data
```

```python
'''
2、知友全国地域分布TOP20（根据人口）
'''
#标准化
def standarize(series_data):
    s = (series_data-series_data.min())/(series_data.max()-series_data.min())*100
    return s

'''人口TOP20'''
habitat_data_series = data["居住地"].value_counts()
#做标准化变化，标准化之前要去掉缺失值
habitat_data_stand = standarize(habitat_data_series[1:])
#habitat_data_stand[0:20].plot(kind = "bar")

'''人口占有率TOP20'''
population_data = pd.read_csv("六普常住人口数.csv",engine = 'python')
popu_city_index = []
#要去掉最后的合计
for s in population_data["地区"][:-1]:
    popu_city_index.append(s[:-1])
#创建新的人口普查序列，index是城市，value为人口
popu_city_series = pd.Series(list(population_data["常住人口"][:-1]),index=popu_city_index)
#将知乎网友地域数量与各城市常住人口数量的Series联合起来
habitat_ratio = habitat_data_series/popu_city_series
habitat_ratio_stand = standarize(habitat_ratio)
habitat_ratio_stand.sort_values(ascending=False)[0:20].plot(kind="bar")
```

```python
'''
3、知友全国地域分布情况（根据学校、粉丝数、关注人数）
'''
import matplotlib.pyplot as plt

fans_data = data[["教育经历","关注者","关注"]]
#计算各大学总共的粉丝和关注者
fans_data_sum = fans_data.groupby("教育经历").sum().drop(['缺失数据','五道口男子职业技术学院','医学','为往圣继绝学',
                                 '我的老師，是山川和大地','重庆第一工程尸培养基地','本科','Phd Candidate'])
fans_data_sum_top = fans_data_sum.sort_values(by="关注者",ascending=False)[0:20]
x = fans_data_sum_top['关注']
y = fans_data_sum_top['关注者']
x_aver = x.mean()
y_aver = y.mean()
plt.figure(figsize=(10,10))
plt.scatter(x,y,s = y/1000,cmap='Blues',c = x,label = '学校')
plt.axvline(x_aver,label='平均关注人数{}'.format(x_aver),color = "r",linestyle='--')
plt.axhline(y_aver,label='平均粉丝人数{}'.format(y_aver),color = 'green',linestyle='--')
plt.legend(loc='upper left')
for i,j,n in zip(x,y,fans_data_sum_top.index):
    plt.text(i+1000,j,n)
```
